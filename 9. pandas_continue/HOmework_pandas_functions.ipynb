```python
#Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
#- оценка 2 и меньше - низкий рейтинг
#- оценка 4 и меньше - средний рейтинг
#- оценка 4.5 и 5 - высокий рейтинг
#Результат классификации запишите в столбец class

import pandas as pd
```

```python
#задание 1
df = pd.read_csv("ratings.csv")

def show_rating(rate):
    if rate['rating'] < 2:
        return 'низкий'
    elif 2 <= rate['rating'] <=4:
        return 'средний'
    elif 4 < rate['rating']:
        return 'высокий'
    
df['rt'] = df.apply(show_rating, axis=1)
df
```

```python
#задание 2
df = pd.read_csv("keywords.csv")
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

reg_dict = {}

for reg in geo_data:
    cities_list = geo_data[reg]
    reg_dict.update({key:value for key,value in zip(cities_list,
                                [reg for i in cities_list])})
print(('sdf sdf sdf').split(' '))
    

def checking(go):
    spl = go['keyword'].split(' ')
    for i in spl:
        if i in reg_dict.keys():
            return reg_dict.get(i)
    
df['region'] = df.apply(checking, axis=1).fillna('undefined')
df.region.value_counts()
```
